```python
def translate(value,leftMin,leftMax,rightMin,rightMax):
  #Figure out how'wide'each range is
  leftSpan=leftMax-leftMin
  rightSpan=rightMax-rightMin
  #Convert the left range intoa0-1 range(float)
  valueScaled=float(value-leftMin)/float(leftSpan)
  #Convert the 0-1 range intoavalue in the right range.
  return rightMin+(valueScaled*rightSpan)
```

```python
import time
import json
import pandas as pd
pd.set_option('display.max_columns', 30)
import matplotlib.pyplot as plt
import paho.mqtt.client as paho
file="RD2.csv"
dataset = pd.read_csv(file)
temp_speed=dataset["Speed (OBD)(km/h)"].tolist()
speed=[]
for i in temp_speed:
    speed.append(translate(i,0,120,20,100))
broker="broker.emqx.io"
details={
"Motor": "DC Motor",
"Acc": "0",
"F/R": "Forward",
"break": "False",
"key": "True"
}
#define callback
def on_message(client, userdata, message):
    #time.sleep(1)
    print("received message =",str(message.payload.decode("utf-8")))

for i in range(len(speed)):
    data=speed[i]
    details["Acc"]=str(int(data))
    temp=json.dumps(details)#string
    
    #print(type(temp))
    client= paho.Client("publish_data")
    client.on_message=on_message
    print("connecting to broker ",broker)
    client.connect(broker)#connect
    client.loop_start() 
    print("subscribing ")
    client.subscribe("/quantanics/ev/substopic/fcf5c4ed824")#subscribe
    #time.sleep(2)
    print("publishing ")
    client.publish("/quantanics/ev/substopic/fcf5c4ed824",temp)#publish
    client.disconnect() #disconnect
    client.loop_stop() #stop loop
```
